### $\epsilon$ for small $r_{jk}$ 

Try plotting the simulated and analytical $\epsilon$ for small $r_{jk}$ again. Now with the new formula:

$$\epsilon(r_{ib}, r_{jk}) = 1 - \frac{m_j e^{-\gamma r_{ij}} + m_k e^{-\gamma r_{ik}}}{m_b e^{-\gamma r_{ib}}},$$

where $r_{ij} \approx r_{ik} \approx \sqrt{r_{ib}^2 + (\frac{r_{jk}}{2})^2}$. For $m_b = m_j + m_k$, the form simplifies to

$$\epsilon(r_{ib}, r_{jk}) = 1 - e^{- \gamma (r_{ij} - r_{ib})},$$

which is a really nice functional form. Can see the error is scaling with the difference between $r_{ij}$ and $r_{ib}$ which makes sense. For the *power law* gravity model, the same result comes out as:

$$\epsilon(r_{ib}, r_{jk}) = 1 - \frac{r_{ij}^{-\gamma}}{r_{ib}^{-\gamma}}.$$

Plotting first with the exponential form:

```python
epsChangeY(0.0001,0.04, 0.3, 50, 1000, ib=True, analytical=True, gamma=2, exp=True)
```

Looks *very* nice, although not sure why it holds so well for large $r_{jk}$, this is perhaps slightly worrying. Better agreement for large $\gamma$ also, why?

```python
epsChangeX(0.1,0.5, 0.02, 100, 1000, ib=True, analytical=True, gamma=2, exp=True)
```

Doesn't work for small $r_{ib}$ since the angle $\delta \theta$ becomes too large? But why isn't this the case for large $r_{jk}$?
