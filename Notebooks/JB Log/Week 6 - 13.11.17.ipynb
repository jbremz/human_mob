## New and improved analytical result for $\epsilon$

For the exponential decay we found:

$$\epsilon(r_{ib}, r_{jk})_{e} = 1 - \frac{m_j e^{-\gamma r_{ij}} + m_k e^{-\gamma r_{ik}}}{m_b e^{- \gamma r_{ib}}},$$

where $r_{ik} \approx r_{ij} \approx \sqrt{r_{ib}^2 + (\frac{r_{jk}}{2})^2}$. For the power law decay:

$$\epsilon(r_{ib}, r_{jk})_{p} = 1 - \frac{m_j r_{ij}^{-\gamma} + m_k r_{ik}^{-\gamma}}{m_b r_{ib}^{-\gamma}}.$$

For $m_b = m_j + m_k$ and $r_{ij} \approx r_{ik}$, these simplify to:

$$\epsilon(r_{ib}, r_{jk})_{e} = 1 - e^{- \gamma (r_{ij}-r_{ib})}$$

and

$$\epsilon(r_{ib}, r_{jk})_{p} = 1 - \bigg(\frac{r_{ij}}{r_{ib}}\bigg)^{- \gamma}.$$

We can easily see the dependence on discrepancy between $r_{ij}$ and $r_{ib}$ as one might expect. $\epsilon(\gamma=0) = 0$ which is reassuring as there is no distance dependence.

## $\tilde{m_b}$ correction with analytical form of $\epsilon$

Have derived the analytical form of the $\tilde{m_b}$ correction for $\epsilon$ in the exponential case.

$$\epsilon(r_{ib}, r_{jk})_{e} = 1 - \frac{e^{- \gamma (r_{ij}-r_{ib})}}{1-k_{jk}e^{-\gamma r_{jk}}},$$

$$k_{jk}^{-1} = 2 \pi \rho \bigg(\frac{e^{-\gamma r_{min}}(\gamma r_{min} + 1) - e^{-\gamma r_{max}}(\gamma r_{max} + 1)}{\gamma^2}\bigg),$$
